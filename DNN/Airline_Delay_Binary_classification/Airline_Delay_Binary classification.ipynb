Importing the dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
import keras
```

Data Collection and Pre-Processing

```python
# loading the data from the csv file to apandas dataframe
df=pd.read_csv('Airline_Delay_Cause.csv')
```

```python
# getting more information about the dataset
df.info()
```

```python
# checking for missing values in each column
df.isnull().sum()
```

```python
# Dropping unnecessary columns: 'carrier', 'carrier_name', 'airport', and 'airport_name' from the DataFrame
df = df.drop(['carrier','carrier_name','airport','airport_name'],axis=1)
df
```

```python
# Dropping all rows with missing values from the DataFrame
df.dropna(inplace=True)
```

```python
df.info()
```

```python
#Getting the minimum and maximum values from the 'weather_delay' column
df['weather_delay'].min(),df['weather_delay'].max()
```

```python
# Creating a new column 'WDCase' where the value is 1 if 'weather_delay' is greater than 100, otherwise 0
df['WDCase']=df['weather_delay'].apply(lambda x:1 if x>100 else 0)
```

```python
# Counting the occurrences of each unique value in the 'WDCase' column
df['WDCase'].value_counts()
```

```python
df
```

Data Pre-Processing

Separating the features & Target

```python
X=df.drop('WDCase',axis=1).values
y=df['WDCase'].values
```

```python
#Splitting the data into training and testing sets with 25% for testing, shuffling the data, and using a random state for reproducibility
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.25, random_state=44, shuffle=True)
```

```python
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

```python
# Defining a Sequential model in Keras
KerasModel = keras.models.Sequential([

    # First hidden layer with 8 neurons, ReLU activation, and no dropout
    keras.layers.Dense(8, activation='tanh'),
    
    # Dropout layer with 10% rate, used to prevent overfitting
    #keras.layers.Dropout(0.1),
    
    # Second hidden layer with 128 neurons and ReLU activation
    keras.layers.Dense(128, activation='sigmoid'),
    
    # Dropout layer with 20% rate, used to prevent overfitting
    #keras.layers.Dropout(0.2),
    
    # Third hidden layer with 64 neurons and ReLU activation
    keras.layers.Dense(64, activation='tanh'),
    
    # Dropout layer with 20% rate, used to prevent overfitting
    #keras.layers.Dropout(0.2),
    
    # Fourth hidden layer with 32 neurons and ReLU activation
    keras.layers.Dense(32, activation='tanh'),
    
    # Dropout layer with 10% rate, used to prevent overfitting
    keras.layers.Dropout(0.2),
    
    # Output layer with 1 neuron and sigmoid activation for binary classification
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
MyOptimizer = tf.keras.optimizers.AdamW(  
    # Initialize the AdamW optimizer
    learning_rate=0.001,  # Set the learning rate to 0.001 (controls step size in updates)
    
    weight_decay=0.004,   # Apply weight decay (L2 regularization) to prevent overfitting
    
    beta_1=0.9,           # Set beta_1, the decay rate for the first moment estimates (momentum)
    
    beta_2=0.999,         # Set beta_2, the decay rate for the second moment estimates (controls variance)
    
    epsilon=1e-07,        # Small constant to avoid division by zero in updates
    
    amsgrad=False,        # Do not apply AMSGrad variant of Adam (use standard AdamW behavior)
    
    name="AdamW"          # Name the optimizer instance as "AdamW"
)
```

```python
 # Compile the Keras model with the specified optimizer, loss function, and evaluation metric
KerasModel.compile(optimizer =MyOptimizer,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Start training the Keras model
history = KerasModel.fit(
    X_train, y_train,                        # Training data: features (X_train) and labels (y_train)
    validation_data=(X_test, y_test),        # Validation data to monitor performance during training
    epochs=100,                              # The model will train for 100 iterations (epochs)
    batch_size=10000,                        # Number of samples to process at once in each iteration
    verbose=1,                               # Display detailed progress of training in the console
    callbacks=[tf.keras.callbacks.EarlyStopping(  # Callback to stop training early if no improvement
        patience=10,                         # Wait for 10 epochs without improvement before stopping
        monitor='val_accuracy',              # Monitor the validation accuracy for improvements
        restore_best_weights=True            # Restore the model to the best weights when stopping
    )]
)
```

```python
print(KerasModel.summary())
```

```python
# Use the trained model to predict labels for the test data (X_test)
y_pred=KerasModel.predict(X_test)
```

```python
print('Prediction Shape is {}'.format(y_pred.shape))
```

```python
print('Prediction items are {}'.format(y_pred[:5]))
```

```python
# Evaluate the model's performance on the test data
ModelLoss, ModelAccuracy = KerasModel.evaluate(X_test, y_test)  

# Print the loss (error) of the model
print('Model Loss is {}'.format(ModelLoss))

# Print the accuracy of the model
print('Model Accuracy is {}'.format(ModelAccuracy))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
# "Loss"
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
y_pred = [np.round(i[0]) for i in y_pred]
CM = confusion_matrix(y_test, y_pred)
print('Confusion Matrix is : \n', CM)
```

```python
# drawing confusion matrix
sns.heatmap(CM, center = True,cmap='Blues_r')
plt.show()
```

```python
ClassificationReport = classification_report(y_test,y_pred)
print('Classification Report is : \n', ClassificationReport )
```
